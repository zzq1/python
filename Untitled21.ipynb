```python
# -*- coding:utf-8 -*-
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
plt.rcParams['axes.unicode_minus']=False
%matplotlib inline
import pandas as pd
from sklearn.cluster import  KMeans
data = pd.read_csv("C:/Users/zzq/Desktop/DataMining/Data/data6/trip.csv")
X = data.iloc[:,2:]
data.head()
X.head()
```

```python
model = KMeans(n_clusters = 3).fit(X)
r1 = pd.Series(model.labels_).value_counts() 
r2 = pd.DataFrame(model.cluster_centers_)
data["jllable "] = model.labels_
data.to_csv("C:/Users/zzq/Desktop/DataMining/Data/data6/new_df.csv",index = False,header = True)
res = data.iloc[:,2:18]

print r1
print r2
res
```

```python
pca = PCA(n_components = 2)
newX = pca.fit_transform(r2)
newX  = pd.DataFrame(newX)
res_new = pca.fit_transform(res)
res_new = pd.DataFrame(res_new)
res_new.columns=["x","y"]
res_new["label"] = model.labels_
res_new.head()
```

```python

plt.rcParams['font.sans-serif'] = ['SimHei']  
plt.plot([res_new[res_new['label']== 0].y], [res_new[res_new['label']== 0].x] ,'ro')
plt.plot([res_new[res_new['label']== 1].y], [res_new[res_new['label']== 1].x] ,'bo')
plt.plot([res_new[res_new['label']== 2].y], [res_new[res_new['label']== 2].x] ,'yo')
plt.rc('font', family='STXihei', size=15)
plt.title(u'聚类图')
plt.xlabel(u'y')
plt.ylabel(u'x')
plt.grid(color='#95a5a6',linestyle='--', linewidth=3,axis='both',alpha=0.4)
plt.savefig(u"C:\\Users\\zzq\\Desktop\\聚类图.png")
plt.show()
```
