```python
# -*- coding:utf-8 -*-
import pandas as pd
import numpy as np
user = pd.read_csv("C:/Users/zzq/Desktop/DataMining/data/fresh_comp_offline/new_train_user.csv")
user.head()
```

```python
item = pd.read_csv("C:/Users/zzq/Desktop/DataMining/data/fresh_comp_offline/tianchi_fresh_comp_train_item.csv")
print item.shape
item.head()
```

```python
#获取唯一的商品ID和商品分类标识
new_item=item.drop_duplicates(['item_id','item_category'])
print new_item.shape
new_item.head()
```

```python
#取出有交互的商品ID和商品分类标识
conn=pd.merge(user,new_item,how='inner',on=['item_id','item_category'])
print conn.shape 
conn.head()
```

```python
#构建训练集,提取12-14到12-16日的item-user pair特征
import numpy as np
dti = pd.date_range(start='20141214',end='20141216')
pydate_array = dti.to_pydatetime()
date_only_array = np.vectorize(lambda s: s.strftime('%Y-%m-%d'))(pydate_array )
date_only_series = pd.Series(date_only_array)
print type(date_only_series)
brows = conn[(conn["behavior_type"] == 1) & (conn["date"].isin(date_only_series))].iloc[:,[0,1,2]].groupby(['user_id','item_id'],as_index = False)["behavior_type"].agg({'brows_count':np.size})
print(brows.shape)
brows.head()
```

```python
#收藏
collect = conn[(conn["behavior_type"] == 2) & (conn["date"].isin(date_only_series))].iloc[:,[0,1,2]].groupby(['user_id','item_id'],as_index = False)["behavior_type"].agg({'collect_count':np.size})
print(collect.shape)
collect.head()
```

```python
#加购
add = conn[(conn["behavior_type"] == 3) & (conn["date"].isin(date_only_series))].iloc[:,[0,1,2]].groupby(['user_id','item_id'],as_index = False)["behavior_type"].agg({'add_count':np.size})
print(add.shape)
add.head()
```

```python
#购买
buy = conn[(conn["behavior_type"] == 4) & (conn["date"].isin(date_only_series))].iloc[:,[0,1,2]].groupby(['user_id','item_id'],as_index = False)["behavior_type"].agg({'buy_count':np.size})
print(buy.shape)
buy.head()
```

```python
#16日18时之后购买的 
laterbuy = conn[(conn["behavior_type"] == 4) & (conn["date"] == "2014-12-16") & (conn["hour"] >= 18)]
laterbuy = laterbuy.copy()
laterbuy["Cartbuytoday"] = 1
print(laterbuy.shape)
laterbuy.head()
```

```python
#16日18时之后加购的
latercart = conn[(conn["behavior_type"] == 3) & (conn["date"] == "2014-12-16") & (conn["hour"] >= 18)]
print(latercart.shape)
latercart.head()
```

```python
#取出16日18时之后加入购物车并且购买的
Cartbuytoday  = pd.merge(latercart,laterbuy,how = "left",on = ['user_id','item_id']).loc[:,["user_id","item_id","Cartbuytoday"]]
Cartbuytoday =Cartbuytoday.fillna(0)
Cartbuytoday = Cartbuytoday[Cartbuytoday['Cartbuytoday'] == 1]
print Cartbuytoday.shape[0]
Cartbuytoday.head()
```

```python
#该用户17日是否购买该商品
target = conn[(conn["behavior_type"] == 4) & (conn["date"] == "2014-12-17")].iloc[:,[0,1]]
target = target.copy()
target['Catrgory'] = 1
print(target.shape)
target.head()
```

```python
#商品浏览数、收藏数、加购数、购买数合并
b_c = pd.merge(brows,collect,how = "outer", on = ['user_id','item_id'])
b_c_a = pd.merge(b_c,add,how = "outer", on = ['user_id','item_id'])
b_c_a_b = pd.merge(b_c_a,buy,how = "outer", on = ['user_id','item_id'])
b_c_a_b_c = pd.merge(b_c_a_b,Cartbuytoday,how = "outer", on = ['user_id','item_id'])
b_c_a_b_c = b_c_a_b_c.fillna(0)
print b_c_a_b_c.shape
b_c_a_b_c.head()

#b_c_a_b_c_t.to_csv("C:/Users/zzq/Desktop/DataMining/b.csv")
```

```python
#12-14到12-16号商品的购买数
tmp_buy = tmp[tmp["behavior_type"] == 4].groupby(['item_id','behavior_type'],as_index = False)["item_id"].agg({'buy_count':np.size}).iloc[:,[0,2]]
tmp_buy.head()
```

```python
#7-16号商品购买数和浏览数合并
from __future__ import division
import numpy as np
brower_buy = pd.merge(tmp_brower,tmp_buy,how="inner",on="item_id")
brower_buy = brower_buy.fillna(0)
brower_buy = brower_buy[(brower_buy["brower_count"] > 0)]
brower_buy = brower_buy.copy()
#购买量/浏览量
brower_buy["buy/brower"] = brower_buy["buy_count"] / brower_buy["brower_count"]
brower_buy = brower_buy.iloc[:,[0,3]]
#brower_buy.to_csv("C:/Users/zzq/Desktop/DataMining/brower_buy.csv")
brower_buy.tail()
```
