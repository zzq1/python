```python
item = pd.read_csv("C:/Users/zzq/Desktop/DataMining/Data/fresh_comp_offline/tianchi_fresh_comp_train_item.csv")
print item.shape
item.head()
```

```python
#new_train_user表中（唯一）用户
unique_user = set(user.user_id)
print "new_train_user表中（唯一）用户:" , len(unique_user)
#new_train_user表中（唯一）商品
unique_user_item = set(user.item_id)
print "new_train_user表中（唯一）商品:" , len(unique_user_item)
#new_train_user表中（唯一的用户商品对数)
unique_user_item_count = user.drop_duplicates(['user_id','item_id'])
print type(unique_user_item_count)
print "new_train_user表中（唯一的用户商品对数):" , len(unique_user_item_count)
```

```python
#train_item表中唯一的商品
unique_item = set(item.item_id)
print "train_item表中唯一的商品:" , len(unique_item)
```

```python
#交互商品数
connect_item_count = unique_user_item & unique_item
print "交互商品数：" , len(connect_item_count)
#没交互商品数
non_connect_item_count = len(unique_user_item) - len(unique_item)
print "交互商品数：" , non_connect_item_count
```

```python
#获取唯一的商品ID和商品分类标识
new_train_item=item.drop_duplicates(['item_id','item_category'])
print new_train_item.shape
new_train_item.head()
```

```python
#取出有交互的商品ID和商品分类标识
sub_trian=pd.merge(new_train_item,user,how='inner',on=['item_id','item_category'])
print sub_trian.shape 
sub_trian.head()
```

```python
#18日真实购买的
true_item = sub_trian[(sub_trian.behavior_type == 4) & (sub_trian.date == "2014-12-18")]
true_item = true_item.copy()
true_item['18日_buy'] = 1
print true_item.shape[0]
true_item.head()
```

```python
#17日18时之后加入购物车的 
predict_item = sub_trian[(sub_trian.behavior_type == 3) & (sub_trian.date == "2014-12-17")& (sub_trian.hour >= 18)]
print predict_item.shape[0]
predict_item.head()
```

```python
#17日18时之后购买的 
sub_trian_tmp = sub_trian[(sub_trian.behavior_type == 4) & (sub_trian.date == "2014-12-17") & (sub_trian.hour >= 18)]
sub_trian_tmp = sub_trian_tmp.copy()
sub_trian_tmp["18时_later_buy"] = 1
print sub_trian_tmp.shape[0]
sub_trian_tmp.head()
```

```python
#取出18时之后加入购物车但没有购买的
get_merge = pd.merge(predict_item,sub_trian_tmp,how = "left",on = ['user_id','item_id']) 
get_merge=get_merge.fillna(0)
get_merge = get_merge[get_merge['18时_later_buy'] == 0]
print get_merge.shape[0]
get_merge.head()
```

```python
#取交集
connection = pd.merge(get_merge, true_item, how="left", on = ["item_id", "user_id"])
connection = connection.fillna(0)
connection = connection[(connection["18日_buy"] == 1) & (connection["18时_later_buy"] == 0)]
print connection.shape[0]
connection.head()
```

```python
from __future__ import division
PredictionSet = len(get_merge)
ReferenceSet = len(true_item)
ConnectionSet = len(connection)
Precision = ConnectionSet / PredictionSet
Recall = ConnectionSet / ReferenceSet
print "F1 = " , 2 * Precision * Recall / (Precision + Recall)
```
