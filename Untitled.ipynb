```python
# -*- coding:utf-8 -*-
#（1）获取数据并进行数据预处理，将含有缺失值的样本去掉，取出死亡率在0<q<=1范围内的数据
import pandas as pd
%matplotlib inline
df=pd.read_csv("C:/Users/zzq/Desktop/DataMining/Data/death rate.csv")
df = df[(df.q_female>0) & (df.q_female<=1) & (df.q_male>0) & (df.q_male<=1)]
df.head()
```

```python
#（2）完成数据的描述和研究对象的分布类型中的分析过程，利用python绘制相关的散点图等。
import matplotlib.pyplot as plt
import numpy as np
#图表字体为华文细黑，字号为15
plt.rc('font', family='STXihei', size=15)
l_x=df[['Age']]
l_y=df[['q_male']]
plt.scatter(l_x,np.log(l_y),60,color='white',marker='.',edgecolors='b',linewidth=3,alpha=0.8)
plt.xlabel(u'年龄')
plt.ylabel(u'男性死亡率')
plt.title(u'年龄与男性死亡率（对数）的关系', family='STXihei', size=20)
#设置背景网格线的颜色，样式，尺寸和透明度
plt.grid(color='#95a5a6',linestyle='--', linewidth=3,axis='both',alpha=0.4)

#保存--到指定路径
plt.gcf().savefig(u'C:/Users/zzq/Desktop/DataMining/第一次实训/年龄与男性死亡率（对数）的关系.png')
plt.show()
plt.close()
```

```python
#（3）根据分析结果，利用年龄、年份和对数生存人口数这3个变量与死亡人数之间的关系，利用普通最小二
#乘线性模型进行模型拟合，并进行预测；将观测值与拟合值，进行图形展示（以样本的观测值做为x轴，拟
#合值作为y轴绘制散点图） 。
from sklearn import linear_model
from sklearn.cross_validation import train_test_split 
import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt
Xi = df.loc[:,['Age','Year','L_male_exp']]
Yi = df.Male_death
train_X,test_X, train_y, test_y = train_test_split(Xi,Yi,test_size = 0.25,random_state = 0)
#设置模型
model = linear_model.LinearRegression()
#训练数据
model.fit(train_X, train_y)
#用训练得出的模型预测数据
y_plot = model.predict(test_X)

#绘图
#plt.scatter(Xi, Yi, color='red',label=u"样本数据",linewidth=2)
plt.plot(test_y, y_plot,'o', color='blue',linewidth=2)
#plt.legend(loc='upper left')
plt.xlabel("Observations")
plt.ylabel("FittingValue")
#plt.title(u"年龄与男性死亡人数的拟合曲线")
plt.gcf().savefig(u'C:/Users/zzq/Desktop/DataMining/年龄与男性对数死亡的拟合曲线.png')
#打印线性方程的权重
#print(model.coef_)
#打印R放
print "R方:", model.score(test_X,test_y)
#打印MSE
#print sp.sqrt(sp.mean((test_y - y_plot) ** 2))
print "MSE:",metrics.mean_squared_error(test_y, y_plot)
plt.show()
plt.close()
```

```python
#（4）再利用多项式回归对数据进行拟合，并进行预测；将观测值与拟合值，进行图形展示（同上）。练习通过调节参数，使拟合达到最优
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge

Xi = df.loc[:,['Age','Year','L_male_exp']]
Yi = df.Male_death
train_X,test_X, train_y, test_y = train_test_split(Xi,Yi,test_size = 0.25,random_state = 0)
print type(train_X)
##这里指定使用岭回归作为基函数
model = make_pipeline(PolynomialFeatures(5), Ridge())
model.fit(train_X, train_y)
##根据模型预测结果
y_plot = model.predict(test_X)
plt.plot(test_y, y_plot, 'o',color='green',linewidth=2)
#plt.legend(loc='upper left')
plt.xlabel("Observations")
plt.ylabel("FittingValue")
#plt.title(u"年龄与男性死亡人数的拟合曲线")
plt.gcf().savefig(u'C:/Users/zzq/Desktop/DataMining/年龄、年份和对数生存人口数与死亡人数之间的关系拟合.png')
#打印R放
print "R方:", model.score(test_X,test_y)
print "MSE:",metrics.mean_squared_error(test_y, y_plot)
plt.show()
```
