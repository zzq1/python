```python
# -*- coding: utf-8 -*-
import pandas as pa
import matplotlib.pyplot as plt
from pandas import *
plt.rcParams['axes.unicode_minus']=False
%matplotlib inline
# legend中显示中文
#plt.legend(prop={'family':'SimHei','size':15}) 
from pylab import *  
plt.rcParams['font.sans-serif'] = ['SimHei']  
df1 = read_csv("scmd_CalculationData.csv")
df1.head()
plt.plot([df1['winddirection']], [df1['windspeed1']],'ro')
plt.rc('font', family='STXihei', size=15)
plt.title(u'风速与风向的函数')
plt.xlabel(u'风向')
plt.ylabel(u'风速')
plt.grid(color='#95a5a6',linestyle='--', linewidth=3,axis='both',alpha=0.4)
plt.savefig(u"C:\\Users\\zzq\\Desktop\\1风速与风向的函数.png")
plt.show()
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.plot(df1['windspeed1'], df1['power'] ,"ro",label=u"平均值")
plt.plot(df1['windspeed1'], df1['power_max'] ,"go",label=u"最大值")
plt.plot(df1['windspeed1'], df1['power_min'] ,"yo",label=u"最小值")
plt.plot(df1['windspeed1'], df1['power_dev'] ,"bo",label=u"标准差")
plt.title(u'功率特性散点图')
plt.grid(color='#95a5a6',linestyle='--', linewidth=3,axis='both',alpha=0.4)
plt.xlabel(u'风速')
plt.ylabel(u'功率')
plt.legend(loc='upper left')
plt.gcf().savefig(u"C:\\Users\\zzq\\Desktop\\2功率特性散点图.png")
plt.show()
```
